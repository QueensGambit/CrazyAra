```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
batch_phases_df = pd.read_csv("batch_phase_data_berlin.csv", header=None).iloc[:,:-1]
```

```python
batch_phases_df
```

```python
batch_phases_df.describe()
```

```python
batches_lists = batch_phases_df.values.tolist()
```

```python
phase0_counts = list()
phase1_counts = list()
phase2_counts = list()
maj_class_counts = list()
for batch_phases in batches_lists[90000:]:
    phase0_count = batch_phases.count(0)
    phase1_count = batch_phases.count(1)
    phase2_count = batch_phases.count(2)
    phase0_counts.append(phase0_count)
    phase1_counts.append(phase1_count)
    phase2_counts.append(phase2_count)
    maj_class_counts.append(max([phase0_count, phase1_count, phase2_count]))
```
